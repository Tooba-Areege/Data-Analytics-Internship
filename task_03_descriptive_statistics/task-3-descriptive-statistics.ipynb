### Step 1: Import library

```python
import pandas as pd
```

### Step 2: Load the cleaned dataset

```python
df = pd.read_csv("cleaned_titanic.csv")
```

### Step 3: Display dataset

```python
df.head()
```

### Step 4: Calculate descriptive statistics

```python
numeric_columns = ["Age", "Fare", "SibSp", "Parch"]

statistics = df[numeric_columns].agg(
    ["count", "mean", "median", "min", "max"]
)

print(statistics)
```

### Step 5: Count categorical values

```python
print("\nGender Counts:")
print(df["Sex"].value_counts())
```

```python
print("\nPassenger Class Counts:")
print(df["Pclass"].value_counts())
```

```python
print("\nSurvival Counts:")
print(df["Survived"].value_counts())
```

### Step 6: Create a clear statistics table

```python
summary = pd.DataFrame({
    "Count": df[numeric_columns].count(),
    "Mean": df[numeric_columns].mean(),
    "Median": df[numeric_columns].median(),
    "Minimum": df[numeric_columns].min(),
    "Maximum": df[numeric_columns].max()
})

print("\nDescriptive Statistics:")
print(summary)
```
